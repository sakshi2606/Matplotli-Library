# MatplotLib 

```python
import matplotlib.pyplot as plt           # to import matplotlib (plt is an alias)
%matplotlib inline                        # to avoid writing plot.show()
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
# Scatter plot

plt.scatter(x,y,c='g')                       # c is color
plt.xlabel('X axis')                         # Naming the x-axis
plt.ylabel('Y axis')                         # Naming the Y-axis
plt.title('My First Scatter plot Graph')     # Giving title to the graph
```

```python
y=x*x
```

```python
# plt plot

plt.plot(x,y,'mo--',linewidth=2,markersize=6)

font={'family': 'Times New Roman',
        'color':  'darkgreen',
        'weight': 'normal',
        'size': 16,
        }

plt.xlabel('X axis',loc='left',fontdict=font)
plt.ylabel('Y axis',loc='bottom',fontdict=font)
plt.title('2-D Graph',loc='center',fontdict=font) 
```

```python
# creating subplots

plt.subplot(2,2,1)             # 2 rows 2 columns and first position
plt.plot(x,y,'b')

plt.subplot(2,2,2)             # 2 rows 2 columns and second position
plt.plot(x,y,'y')

plt.subplot(2,2,3)             # 2 rows 2 columns and third position
plt.plot(x,y,'g')

plt.subplot(2,2,4)             # 2 rows 2 columns and fourth position
plt.plot(x,y,'r')
```

```python
# Bar plot

x=[2,4,8]


plt.bar(x,y,color='y')
```

### Histogram

Histogram are used to plot the density (or frequency) w.r.t whatever we are plotting

```python
a=np.array([22,87,35,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)                # in x-axis you will have all the numbers and in y-axis you will have the densities or frquency
plt.title('Histogram')
```

### Pie Chart


```python
labels= 'Python','SQL','Power BI','Excel'
sizes=[35,27,23,20]
colors=['blue','yellow','green','red']
explode= (0.1,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True, autopct='%1.1f%%')
```
